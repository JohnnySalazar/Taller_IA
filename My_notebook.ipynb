```python
#Se importan la librerias a utilizar
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
from google.colab import files

# Selecciona un archivo desde tu computadora
upload = files.upload()

# Esto imprimirá información sobre el archivo cargado
for filename in upload.keys():
    print(f'Archivo cargado: {filename}')
```

```python
import pandas as pd

# Lee el archivo cargado usando pandas
dataset = pd.read_csv(next(iter(upload)))

# Seleccionar todas las columnas excepto la columna "Class" como atributos en X
X = dataset.drop('Class', axis=1)

# Seleccionar la columna "Class" como etiquetas en y
y = dataset['Class']

# Ver las primeras 5 filas del DataFrame
print(X.head())
print(" ")
print(y.head())
print(" ")

# Obtener el número de filas y columnas de X
num_rows, num_columns = X.shape
print(f"Número de filas de X: {num_rows}")
print(f"Número de columnas de X: {num_columns}")

# Obtener el número de elementos únicos en y
num_unique_labels = y.nunique()
print(f"Número de etiquetas únicas en y: {num_unique_labels}")
```

```python
#Separo los datos de "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
escalar = StandardScaler()
X_train = escalar.fit_transform(X_train)  ## Encontrar la media y varianza
X_test = escalar.transform(X_test)
```

```python
model = LogisticRegression()
#Entrenamos el modelo
model.fit(X_train, y_train)  ##Entranamiento
```

```python
import joblib

# Guardar el modelo en un archivo
model_filename = 'modelo_entrenado.pkl'
joblib.dump(model, model_filename)

print("Modelo guardado en", model_filename)
```

```python
#Realizamos una predicción
y_pred = model.predict(X_test)
```

```python
#Verifico la matriz de Confusión

matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
```

```python
precision = precision_score(y_test, y_pred)
print('Precisión del modelo:')
print(precision)
```

```python
exactitud = accuracy_score(y_test, y_pred)
print('Exactitud del modelo:')
print(exactitud)
```
